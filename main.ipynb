```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()
```

```python
textFile = spark.read.text("README.md")
```

```python
textFile.filter(textFile.value.contains("sdf")).collect()
```

```python
from pyspark.sql.functions import *
```

```python
textFile.select(size(split(textFile.value, "\s+")).name("numWords")).agg(
    max(col("numWords"))
).collect()
```

```python
word_counts = textFile.select(
    explode(split(textFile.value, "\s+")).alias("word")
).groupby("word").count()
```

```python
word_counts.collect()
```

```python
word_counts.cache()
```

```python
word_counts.count()
```
